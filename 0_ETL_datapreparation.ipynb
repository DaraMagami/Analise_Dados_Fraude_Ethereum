# ETL - Transaction Dataset

```python
import sqlite3
import pandas as pd
```

```python
df = pd.read_csv("transaction_dataset.csv", index_col=0) # Lê o arquivo csv e inclui um DataFrame.
```

```python
con = sqlite3.connect("projetorisco.db") # Conecta ao banco de dados SQLite.
```

```python
cursor = con.cursor() # Armazena o cursor.
```

```python
df.to_sql("transaction_dataset", con, if_exists="replace", index=False) # Salva o arquivo csv como tabela do Banco de 
                                                                        # Dados. 
```

```python
pd.set_option('display.max_columns', None) # Estabelece que todas as colunas do DataFrame apareçam.
```

```python
df_geral = pd.read_sql_query("""
    SELECT *
    FROM transaction_dataset
""", con) # Cria um novo DataFrame e armazena a consulta SQL. 

df_geral.head() # Apresenta o DataFrame. 
```

```python
df_geral.info() # Traz o formato dos dados.
```

```python
# Exclui a coluna 'Index'
df_geral = df_geral.drop(columns=['Index'])
```

```python
# Converte a coluna 'FLAG' de int64 para object. Isso é feito para evitar que essas colunas sejam consideradas 
# nas estatísticas como valores numéricos.
df_geral['FLAG'] = df_geral['FLAG'].astype(object)
```

```python
df_geral.to_sql("transaction_dataset", con, if_exists="replace", index=False) # Converte o DataFrame df_geral em uma 
                                                                              # tabela de mesmo nome do DataBase 
                                                                              # projetoecommerce.
```

```python
df_geral.info() # Traz o formato dos dados.
```

```python
df_geral.columns = df_geral.columns.str.strip() # Remove espaços extras.
```

```python
df_geral.columns = df_geral.columns.str.replace(' ', '_') # Substitui espaços por underline.
```

```python
df_geral.to_sql("transaction_dataset", con, if_exists="replace", index=False) # Converte o DataFrame df_geral em uma 
                                                                              # tabela de mesmo nome do DataBase 
                                                                              # projetoecommerce.
```

```python
cursor.close() # Fecha o cursor.

con.close() # Fecha a conexão.
```
